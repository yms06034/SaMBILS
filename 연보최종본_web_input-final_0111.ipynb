```python
def second_one_page_field(browser, all_p, fl, cw, a4):
    # 자영업자,무급가족 종사자
    man_self_employed_input = find_id('A14_1_1', browser)
    girl_self_employed_input = find_id('A14_2_1', browser)

    # 상용근로자 
    man_office_input = find_id('A14_1_2', browser)
    girl_office_input = find_id('A14_2_2', browser)

    # 임시 및 일용근로자
    man_daily_worker_input = find_id('A14_1_3', browser)
    girl_daily_worker_input = find_id('A14_2_3', browser)

    # 기타종사자
    man_other_worker_input = find_id('A14_1_4', browser)
    girl_other_worker_input = find_id('A14_2_4', browser)

    man_self_employed_input.clear()
    girl_self_employed_input.clear()

    man_office_input.clear()
    girl_office_input.clear()

    man_daily_worker_input.clear()
    girl_daily_worker_input.clear()

    man_other_worker_input.clear()
    girl_other_worker_input.clear()

    find_id('A15_1', browser).clear()
    find_id('A15_2', browser).clear()
    find_id('A15_3', browser).clear()

    randint = random.choice([5, 6, 7])
    # 상용근로자
    r = all_p - fl - cw

    r_man = round((r * randint) / 10)
    r_girl =  round(r - r_man)

    fl_man = round((fl * randint) / 10)
    fl_girl = round(fl - fl_man)

    cw_man = round((cw * randint) / 10)
    cw_girl = round(cw - cw_man)

    total = 100
    split_ratio = [r, fl, cw]

    noe_ratio = [total * ratio / sum(split_ratio) for ratio in split_ratio]

    r_money = math.trunc(int(a4 * math.trunc(noe_ratio[0]) / 100))
    fl_money = math.trunc(int(r_money * noe_ratio[1] / 100))
    cw_money = math.trunc(int(a4 - (fl_money + r_money)))

    # 자영업자,무급가족 종사자
    man_self_employed_input.send_keys(0)
    girl_self_employed_input.send_keys(0)
    # 상용근로자 
    man_office_input.send_keys(r_man)
    girl_office_input.send_keys(r_girl)
    # 임시 및 일용근로자
    man_daily_worker_input.send_keys(fl_man)
    girl_daily_worker_input.send_keys(fl_girl)
    # 기타종사자
    man_other_worker_input.send_keys(cw_man)
    girl_other_worker_input.send_keys(cw_girl)
    # 연간급여액 (퇴직금, 복리후생비 제외)
    ## 상용근로자

    find_id('A15_1', browser).send_keys(r_money)
    ## 임시 및 일용근로자
    find_id('A15_2', browser).send_keys(fl_money)
    ## 기타종사자
    find_id('A15_3', browser).send_keys(cw_money)
```
